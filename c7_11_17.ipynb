```python
#Plotting signals

#3 traces
def plt_signal_T3(ch1_amp,ch2_amp,ch3_amp,ch1_t,ch2_t,ch3_t,inj,T1,T2,T3, ch1, ch2, ch3, x1,x2):
  plt.figure(figsize=(6,6));
  plt.ylabel("Amplitude", fontweight="heavy")
  plt.xlabel("Time (sec)", fontweight="heavy")
  plt.xlim(x1,x2)

  plt.title(f"Inject: Trace {inj} for T={T1},{T2},{T3}", fontweight="heavy")
  plt.plot(ch1_t,ch1_amp, label=f"Trace {T1}/Channel {ch1}", color = "blue");
  plt.plot(ch2_t,ch2_amp, label=f"Trace {T2}/Channel {ch2}", color = "green");
  plt.plot(ch3_t,ch3_amp, label=f"Trace {T3}/Channel {ch3}", color = "orange");

  plt.legend(loc='upper left', bbox_to_anchor=(1, 1));
  plt.show()

  #  (|max1|-|max2|)/|max1| = %
  #  (|max1|-|max3|)/|max1| = %
  if ((inj==7) or (inj==11) or (inj==17)):
    if (inj == T1):
      inj_amp = ch1_amp
      abs_inj_amp = np.abs(ch1_amp)
      abs_ch2_amp = np.abs(ch2_amp)
      abs_ch3_amp = np.abs(ch3_amp)

      max_inj = np.max(abs_inj_amp)
      max2 = np.max(abs_ch2_amp)
      max3 = np.max(abs_ch3_amp)
    elif (inj == T2):
      inj_amp = ch2_amp
      abs_ch1_amp = np.abs(ch1_amp)
      abs_inj_amp = np.abs(ch2_amp)
      abs_ch3_amp = np.abs(ch3_amp)

      max_inj = np.max(abs_inj_amp)
      max2 = np.max(abs_ch1_amp)
      max3 = np.max(abs_ch3_amp)
    else:
      inj_amp = ch3_amp
      abs_ch1_amp = np.abs(ch1_amp)
      abs_ch2_amp = np.abs(ch2_amp)
      abs_inj_amp = np.abs(ch3_amp)

      max_inj = np.max(abs_inj_amp)
      max2 = np.max(abs_ch1_amp)
      max3 = np.max(abs_ch2_amp)

    accuracy_inj_ch2 = (max_inj - max2)/max_inj
    accuracy_inj_ch3 = (max_inj - max3)/max_inj

    # average of amplitude
    avg_ch1 = np.mean(ch1_amp)
    avg_ch2 = np.mean(ch2_amp)
    avg_ch3 = np.mean(ch3_amp)

    print(f"Absolute Maximum Value: Injectected: {max_inj:.6f}, Other non-injected: {max2:.6f}, {max3:.6f}.")

    print(f"Accuracy with Trace {inj} and Trace {T2}: {accuracy_inj_ch2*100:.2f}%. {100-accuracy_inj_ch2*100:.2f}% crosstalk.")
    print(f"Accuracy with Trace {inj} and Trace {T3}: {accuracy_inj_ch3*100:.2f}%. {100-accuracy_inj_ch3*100:.2f}% crosstalk.")

    print(f"Average Amplitude for Trace {T1}(Channel {ch1}): {avg_ch1:.6f}.")
    print(f"Average Amplitude for Trace {T2}(Channel {ch2}): {avg_ch2:.6f}.")
    print(f"Average Amplitude for Trace {T3}(Channel {ch3}): {avg_ch3:.6f}.")
```

```python
def rise_time(t_arr,amp_arr, ch,base1,base2,exp1,exp2):
  plt.figure(figsize=(6,6))
  plt.plot(t_arr,amp_arr, label=f"Channel {ch}", color = "red");
  plt.xlim(-base1*10**-exp1,base2*10**-exp2);
  plt.title("Rise Time", fontweight="heavy");
  plt.xlabel("Time (sec)", fontweight="heavy");
  plt.ylabel("Amplitude", fontweight="heavy");
  plt.show()

  rise_time = base1*10**-exp2+base1*10**-exp2
  nano_time = rise_time/(10**-9)
  print(f"Rise time: {rise_time} sec = {nano_time:.6f} nano sec.")
```

```python
def noise_crosstalk(amp_noise,amp_list):
  absmax_amp=(max(abs(amp_list[0][1])))
  print(f"Abs max = {absmax_amp:.6f}\n")

  absmax_noise_list = []
  for i in range(0,3):
    absmax_noise_list.append(max(abs(amp_noise[i])))
    print(f"Abs noise max ch{i+1}= {absmax_noise_list[i]:.6f}")

    noise_crosstalk = (1-(absmax_amp-absmax_noise_list[i])/absmax_amp)*100

    print(f"Noise cross talk ch{i+1} = {noise_crosstalk:.6f}%\n")
```

```python
noise_crosstalk(amp_noise1,amp_list)
```

```python
def position_plot(trace_i,indice,amp_list,ch,color):
  absmax_amp_ch = []
  for i in indice:
    absmax_amp_ch.append(max(abs(amp_list[ch-1][i])))
  if(i==1):
    max_index=i

  max_amp = max(absmax_amp_ch)
  print(f"Max {ch}: {max_amp}")

  plt.figure(figsize=(16,6))
  plt.title(f"Position Plot: Ch {ch}", fontweight="heavy", fontsize=20)
  plt.xlabel("Trace Num", fontweight="heavy")
  plt.ylabel("Max Amplitude", fontweight="heavy")
  plt.xticks(range(6, 23, 1))
  plt.grid(True)

  max_percent = []

  for i in absmax_amp_ch:
    max_curr = 1-(max_amp-i)/max_amp
    max_percent.append(max_curr)

  plt.plot(trace_i,absmax_amp_ch, label=f"Channel {ch}", color = f"{color}",marker="o");
  for i in range(len(absmax_amp_ch)):
      plt.annotate(f'{max_percent[i]*100:.2f}%', (trace_i[i], absmax_amp_ch[i]), fontsize=10, fontweight="heavy")

  plt.legend(loc='upper left', bbox_to_anchor=(1, 1));
  plt.show()

  print(f"Avg Max Amp for Ch1: {np.mean(absmax_amp_ch)}")


#zoomed in
  plt.figure(figsize=(16,6))
  plt.title(f"Position Plot: Ch {ch}", fontweight="heavy", fontsize=20)
  plt.xlabel("Trace Num", fontweight="heavy")
  plt.ylabel("Max Amplitude", fontweight="heavy")
  plt.xticks(range(6, 23, 1))
  plt.grid(True)
  plt.plot(trace_i,absmax_amp_ch, label=f"Channel {ch}", color = f"{color}",marker="o");
  for i in range(len(absmax_amp_ch)):
    plt.annotate(f'{max_percent[i]*100:.2f}%', (trace_i[i], absmax_amp_ch[i]), fontsize=10, fontweight="heavy")
  plt.ylim(0,.055)

  plt.legend(loc='upper left', bbox_to_anchor=(1, 1));

#ch1,2,3
def position_plots(trace_i,indice,amp_list,ch1,ch2,ch3,c1,c2,c3):
  absmax_amp_ch1 = []
  for i in indice:
    absmax_amp_ch1.append(max(abs(amp_list[ch1-1][i])))
  if(i==1):
    max_index=i

  max_amp1 = max(absmax_amp_ch1)
  print(f"Max {ch1}: {max_amp1}")

  absmax_amp_ch2 = []
  for i in indice:
    absmax_amp_ch2.append(max(abs(amp_list[ch2-1][i])))
  if(i==1):
    max_index=i

  max_amp2 = max(absmax_amp_ch2)
  print(f"Max {ch2}: {max_amp2}")

  absmax_amp_ch3 = []
  for i in indice:
    absmax_amp_ch3.append(max(abs(amp_list[ch3-1][i])))
  if(i==1):
    max_index=i

  max_amp3 = max(absmax_amp_ch3)
  print(f"Max {ch3}: {max_amp3}")

  plt.figure(figsize=(16,6))
  plt.title("Position Plot", fontweight="heavy", fontsize=20)
  plt.xlabel("Trace Num", fontweight="heavy")
  plt.ylabel("Max Amplitude", fontweight="heavy")
  plt.xticks(range(6, 23, 1))
  plt.grid(True)

  max_percent1 = []
  max_percent2 = []
  max_percent3 = []

  for i in absmax_amp_ch1:
    max_curr1 = 1-(max_amp1-i)/max_amp1
    max_percent1.append(max_curr1)

  for i in absmax_amp_ch2:
    max_curr2 = 1-(max_amp2-i)/max_amp2
    max_percent2.append(max_curr2)

  for i in absmax_amp_ch3:
    max_curr3 = 1-(max_amp3-i)/max_amp3
    max_percent3.append(max_curr3)

  plt.plot(trace_i,absmax_amp_ch1, label=f"Channel {ch1}", color = f"{c1}",marker="o");
  for i in range(len(absmax_amp_ch1)):
      plt.annotate(f'{max_percent1[i]*100:.2f}%', (trace_i[i], absmax_amp_ch1[i]), fontsize=10, fontweight="heavy")

  plt.plot(trace_i,absmax_amp_ch2, label=f"Channel {ch1}", color = f"{c2}",marker="o");
  for i in range(len(absmax_amp_ch2)):
      plt.annotate(f'{max_percent2[i]*100:.2f}%', (trace_i[i], absmax_amp_ch2[i]), fontsize=10, fontweight="heavy")

  plt.plot(trace_i,absmax_amp_ch3, label=f"Channel {ch1}", color = f"{c3}",marker="o");
  for i in range(len(absmax_amp_ch3)):
      plt.annotate(f'{max_percent3[i]*100:.2f}%', (trace_i[i], absmax_amp_ch3[i]), fontsize=10, fontweight="heavy")

  plt.legend(loc='upper left', bbox_to_anchor=(1, 1));

  plt.show()


  #zoomed
  plt.figure(figsize=(14,6))
  plt.title("Position Plot", fontweight="heavy", fontsize=20)
  plt.xlabel("Trace Num", fontweight="heavy")
  plt.ylabel("Max Amplitude", fontweight="heavy")
  plt.xticks(range(6, 23, 1))
  plt.ylim(0,.055)
  plt.grid(True)

  plt.plot(trace_i,absmax_amp_ch1, label=f"Channel {ch1}", color = f"{c1}",marker="o");
  for i in range(len(absmax_amp_ch1)):
      plt.annotate(f'{max_percent1[i]*100:.2f}%', (trace_i[i], absmax_amp_ch1[i]), fontsize=10, fontweight="heavy")

  plt.plot(trace_i,absmax_amp_ch2, label=f"Channel {ch1}", color = f"{c2}",marker="o");
  for i in range(len(absmax_amp_ch2)):
      plt.annotate(f'{max_percent2[i]*100:.2f}%', (trace_i[i], absmax_amp_ch2[i]), fontsize=10, fontweight="heavy")

  plt.plot(trace_i,absmax_amp_ch3, label=f"Channel {ch1}", color = f"{c3}",marker="o");
  for i in range(len(absmax_amp_ch3)):
      plt.annotate(f'{max_percent3[i]*100:.2f}%', (trace_i[i], absmax_amp_ch3[i]), fontsize=10, fontweight="heavy")

  plt.legend(loc='upper left', bbox_to_anchor=(1, 1));

  print(f"Avg Max Amp for Ch1: {np.mean(absmax_amp_ch1)}")
  print(f"Avg Max Amp for Ch1: {np.mean(absmax_amp_ch2)}")
  print(f"Avg Max Amp for Ch1: {np.mean(absmax_amp_ch3)}")
```
